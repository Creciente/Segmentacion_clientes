# **Modelo K-Means**

```python
from sklearn.cluster import KMeans
```

```python
scaler = StandardScaler()
scaler_df2= scaler.fit_transform(df2)
```

```python
# Visualizar la data
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(projection = '3d')
ax.scatter(df2['Year_Birth'], df2['Kidhome'], df2['Teenhome'])
ax.set_xlabel('Año de Nacimiento')
ax.set_ylabel('Número de Niños')
ax.set_zlabel('Número de Adolescentes')

plt.title("Visualización de Cantidad de Niños y Adolescentes \n por Año de Nacimiento del cliente")
```

```python
# Instanciar Kmeans

kmeans= KMeans(n_clusters=2)
kmeans.fit(scaler_df2)
```

```python
df2['cluster']= kmeans.labels_
```

```python
# Visualización de los clusters
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(projection = '3d')
ax.scatter(df2['Year_Birth'], df2['Kidhome'], df2['Teenhome'], c=df2['cluster'],cmap= 'viridis')
ax.set_xlabel('Año de Nacimiento')
ax.set_ylabel('Número de Niños')
ax.set_zlabel('Número de Adolescentes')
plt.grid(c='g',alpha= 0.8)
plt.title("Visualización de Cantidad de Niños y Adolescentes \n por Año de Nacimiento")
```

```python
# Buscar óptimización de cluster
# Método del codo

inertias = []
for k in range(2,32):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaler_df2)
    inertias.append(kmeans.inertia_)
```

```python
# Visualización de scores

plt.plot(range(2,32), inertias, marker = '.', color= 'green')
plt.xlabel('clusters')
plt.ylabel('inertia')
```

Codo poco visible a simple vista.
Posiblemente cluster= 3, 5 y 9

```python
# Estudio del índice silueta

from sklearn.metrics import silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer

silhouette_scores = []

for i in range(2, 32):
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(scaler_df2)

  silhouette_scores.append(silhouette_score(scaler_df2, kmeans.labels_))
```

```python
plt.plot(range(2, 32), silhouette_scores, marker = '.',color='green')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score');
```

Mejor cluster en 2, pero se descarta porque no cumple el requisito del estudio.
Mejor segundo cluster en 3.
Se opta por el cluster 3

```python
# Modelo para cluster 3
kmeans3 = KMeans(n_clusters = 3)
kmeans3.fit(scaler_df2)

silhouette_score(scaler_df2, kmeans3.labels_)
```

```python
df2['cluster3'] = kmeans3.labels_
```
